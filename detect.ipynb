```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("fraudTest.csv")
df
```

```python
df.info()
```

```python
df.drop(["Unnamed: 0","trans_date_trans_time"], axis=1, inplace=True)
df
```

```python
df['category'].value_counts()
```

```python
df['category_freq']= df['category'].map(df['category'].value_counts())
df.drop('category', axis=1, inplace=True)
```

```python
df["merchant"].value_counts()
```

```python
len(df["merchant"].unique())
```

```python
df['merchant_freq'] = df['merchant'].map(df['merchant'].value_counts())
df.drop('merchant', axis=1, inplace=True)
```

```python
df.head()
```

```python
y = df[["is_fraud"]]
```

```python
df["city"].value_counts()
```

```python
df['city_target'] = df.groupby('city')['is_fraud'].transform('mean')
```

```python
df["state"].value_counts()
```

```python
len(df["state"].value_counts())
```

```python
df['state_freq'] = df['state'].map(df['state'].value_counts())
df.drop('state', axis=1, inplace=True)
```

```python
df["job"].value_counts()
```

```python
df['job_freq'] = df["job"].map(df["job"].value_counts())
df.drop("job", axis=1, inplace= True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df["gender"] = le.fit_transform(df["gender"])
```

```python
x = df[["merchant_freq","category_freq","city_target","state_freq","job_freq"]]
x
```
