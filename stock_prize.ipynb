```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("nestle.csv")
df
```

```python
df.info()
```

```python
df['Deliverable Quantity'] = df['Deliverable Quantity'].fillna(df['Deliverable Quantity'].median())
df['% Deli. Qty to Traded Qty'] = df['% Deli. Qty to Traded Qty'].fillna(df['% Deli. Qty to Traded Qty'].median())
df.head()
```

```python
x = df[["Open Price","High Price","Low Price","No.of Shares","No. of Trades","Deliverable Quantity"]]
```

```python
y = df["Close Price"]
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=43)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_test.iloc[0]
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```
